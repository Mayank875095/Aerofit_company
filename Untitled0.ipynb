**2.Detect Outliers (using boxplot, “describe” method by checking the difference between mean and median)**

```python
print('Median Age is :',np.median(df['Age']),'Mean Age is :',np.round(np.mean(df['Age']),2),'Diffrence :',np.round(np.mean(df['Age']) - np.median(df['Age']),2))
print('Median Age is :',np.median(df['Usage']),'Mean Age is :',np.round(np.mean(df['Usage']),2),'Diffrence :',np.round(np.mean(df['Usage']) - np.median(df['Usage']),2))
print('Median Age is :',np.median(df['Income']),'Mean Age is :',np.round(np.mean(df['Income']),2),'Diffrence :',np.round(np.mean(df['Income']) - np.median(df['Income']),2))
print('Median Age is :',np.median(df['Miles']),'Mean Age is :',np.round(np.mean(df['Miles']),2),'Diffrence :',np.round(np.mean(df['Miles']) - np.median(df['Miles']),2))
```

```python
def describe_2(x):
    return pd.Series([x.mean(),x.median(),x.std(),abs(x.mean()-x.median())],
                    index=['Mean','Median','Standard deviation','Difference btw 2M'])
df_int_float = df.loc[:,(df.dtypes == 'float64') | (df.dtypes == 'int64')]
df_int_float.apply(describe_2)
```

**4 . Representing the marginal probability like - what percent of customers have purchased KP281, KP481, or KP781 in a table (can use pandas.crosstab here) :**

```python
marginal_probability = np.round(pd.crosstab(df['Product'],
                     [df['MaritalStatus'],df['Gender']],
                     margins=True,
                     margins_name='Percent',
                     normalize=True)*100,2)
marginal_probability
```

```python
sns.heatmap(marginal_probability,annot=True,cmap="YlGnBu",square=True,fmt='')
plt.show()
```

***6 . With all the above steps you can answer questions like: What is the probability of a male customer buying a KP781 treadmill? :***

```python
# Probability of a male customer buying a KP781 treadmill
male_prob = len(df[(df['Gender'] == 'Male') & (df['Product'] == 'KP781')]) / len(df) * 100
male_prob_text = f"Probability of a male customer buying a KP781 treadmill: {male_prob:.2f}%"
male_prob_text
```

```python
Male_KP781 = df[(df['Gender'] == 'Male') & (df['Product'] == 'KP781')]
np.round(pd.crosstab(Male_KP781['Product'],
            Male_KP781['Gender'])/len(df) * 100,2)
```

```python
# Probability of a women customer buying a KP781 treadmill
female_prob = len(df[(df['Gender'] == 'Female') & (df['Product'] == 'KP781')]) / len(df) * 100
female_prob_text = f"Probability of a female customer buying a KP781 treadmill: {female_prob:.2f}%"
female_prob_text
```
